### Importing libraries

```python
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

### Check on the daily activity data 

```python
daily_df = pd.read_csv("HealthTrackerData/dailyActivity_summary.csv")
daily_df.head()
```

```python
num_user = len(set(daily_df.Id))
f"Total unqiue users: {num_user}"
```

```python
daily_df.groupby(["Id"]).agg({"ActivityDate": "count"}).head(num_user)
```

### Check on the hourly activity data

```python
hourly_df = pd.read_csv("HealthTrackerData/hourlyActivity_summary.csv")
hourly_df.head()
```

```python
num_user = len(set(hourly_df.Id))
f"Total unqiue users: {num_user}"
```

```python
hourly_df.groupby(["Id"]).agg({"ActivityHour": "count"}).head(num_user)
```

```python
diff_in_calories = hourly_df.Calories.sum() - daily_df.Calories.sum() 
f"Differences in calories between daily and hourly count: {diff_in_calories}"
```
